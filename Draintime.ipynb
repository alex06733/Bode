```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
time_values_7rpm = np.array([40.4, 80.8, 121.2, 161.6, 202, 242.4, 282.8, 323.2, 363.6, 404])  # Sekunden
time_values_15rpm = np.array([18.85, 37.7, 56.55, 75.4, 94.25, 113.1, 131.95, 150.8, 169.65, 188.5])  # Sekunden
volume_values = np.arange(100, 1100, 100)  # mL
time_errors = 0.3  # Sekunden
volume_errors = 4.33  # mL

# Lineare Regression
coefficients_7rpm = np.polyfit(time_values_7rpm, volume_values, 1)
linear_fit_7rpm = np.poly1d(coefficients_7rpm)

coefficients_15rpm = np.polyfit(time_values_15rpm, volume_values, 1)
linear_fit_15rpm = np.poly1d(coefficients_15rpm)

# Plot
plt.figure(figsize=(10, 6))

# Fehlerbalken 7 U/min
plt.errorbar(
    time_values_7rpm,
    volume_values,
    xerr=time_errors,
    yerr=volume_errors,
    fmt='o',
    color='red',
    capsize=2,
    label='7 U/min Datenpunkte'
)

# Fehlerbalken 15 U/min
plt.errorbar(
    time_values_15rpm,
    volume_values,
    xerr=time_errors,
    yerr=volume_errors,
    fmt='o',
    color='blue',
    capsize=2,
    label='15 U/min Datenpunkte'
)

# Lineare Anpassungen
plt.plot(time_values_7rpm, linear_fit_7rpm(time_values_7rpm), linestyle='--', color='red', label='7 U/min Fit')
plt.plot(time_values_15rpm, linear_fit_15rpm(time_values_15rpm), linestyle='--', color='blue', label='15 U/min Fit')

# Achsenbeschriftung
plt.xlabel("Ablaufzeit des Esters (s)")
plt.ylabel("Ablaufvolumen (mL)")

# Titel und Gitter
plt.title("Ablaufzeit des Esters über der Drehzahl der Exzenterschneckenpumpe")
plt.grid(True, linestyle='--', alpha=0.7)

# Legende
plt.legend()

# Plot anzeigen
plt.show()
```
